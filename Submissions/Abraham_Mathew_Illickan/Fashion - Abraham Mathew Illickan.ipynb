```python
import pandas as pd
fp = 'C:\\Users\\ami12\\Downloads\\fashion-mnist_train.csv'
```

```python
data = pd.read_csv(fp)
data.describe()
```

```python
y = data['label']
```

```python
y.head()
```

```python
x=data.drop(['label'],axis=1)
```

```python
x.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=7456)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(random_state=1)
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
rep = classification_report(y_test,model.predict(X_test))
```

```python
print(rep)
```

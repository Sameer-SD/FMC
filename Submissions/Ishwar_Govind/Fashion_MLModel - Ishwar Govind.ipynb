```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
train_data = pd.read_csv('C:\\Users\\ishwa\\Desktop\\work\\FMC\\fashion-mnist_train\\fashion-mnist_train.csv')
train_data.head()
```

```python
y = train_data.label
train_new = train_data.drop(['label'],axis=1)
x = train_new[train_new.columns]
x.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
model = RandomForestClassifier(n_estimators=100,random_state=42,criterion = 'entropy')
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,model.predict(x_test)))
```

```python
final_test_data = pd.read_csv('C:\\Users\\ishwa\\Desktop\\work\\FMC\\fashion-mnist_train\\Q2_Clothing_test.csv')
final_x = final_test_data
final_pred = model.predict(final_x)
```

```python
output = pd.DataFrame({'label' : model.predict(final_x) })
output.to_csv('my_submission.csv', index = False)
```

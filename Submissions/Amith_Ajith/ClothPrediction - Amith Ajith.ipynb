```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("fashion-mnist_train.csv") #please give the correct name
```

```python
data.head()
```

```python
data.shape
```

```python
type=['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle Boot']
```

```python
train_x =data.drop(['label'],axis=1)
train_y= data['label']
train_x=(np.array(train_x)).astype('float32')
train_y=np.array(train_y)
```

```python
train_x,train_y
```

```python
train_x.shape
```

```python
train_x=train_x/255
```

```python
train_X=train_x.reshape(train_x.shape[0],28,28)
```

```python
train_X
```

```python
train_y
```

```python
train_X.shape
```

```python
train_y.shape
```

```python
len(train_y)
```

```python
plt.figure()
plt.imshow(train_X[0])
plt.colorbar()
plt.grid(False)
plt.show
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(train_X[i],cmap=plt.cm.binary)
    plt.xlabel(type[train_y[i]])
    
plt.show()
```

```python
train_y
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128,activation=tf.nn.relu),
     keras.layers.Dense(128,activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_X,train_y,epochs=60)#please bear with it 
```

```python
test=pd.read_csv('Q2_Clothing_test.csv')#please give the correct name
```

```python
test.head()
```

```python
test_x=np.array(test)
```

```python
test_x.shape
```

```python
test_x=test_x/255
```

```python
test_X=test_x.reshape(test_x.shape[0],28,28)
```

```python
pred=model.predict(test_X)
```

```python
pred
```

```python
np.argmax(pred[0])
```

```python
plt.figure()
plt.imshow(test_X[0])
plt.colorbar()
plt.grid(False)
plt.show
```

```python
plt.figure(figsize=(20,20))
for i in range(100):
    plt.subplot(10,10,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(test_X[i],cmap=plt.cm.binary)
    plt.xlabel(type[np.argmax(pred[i])])
    
plt.show()
```
